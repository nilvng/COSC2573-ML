```python
import math
def entropy(data, target):
        entropy_val = 0
        vCounts = pd.value_counts(data[target])
        # display(vCounts)
        for target_val, total in enumerate(vCounts):
            pi = total / data.shape[0] # probability of one class
            entro = pi * math.log(pi,2)
            entropy_val += entro  
        return -entropy_val
```

```python
entropy(sail_df,"Sail")
```

```python
entropy(zoo_df,"type")
```

```python
def majority_class(data, target):
    majority = 0
    clas = ""
    vCounts = pd.value_counts(data[target])
    # display(vCounts)
    for index, total in enumerate(vCounts):
        if total > majority:
            majority = total
            clas = vCounts.axes[0][index]
    return clas

# or just use idmax()
```

```python
majority_class(sail_df, "Sail")
```

```python
def simple_rule_learner(data, target):
    while data.shape[0] > 0 : # stop finding rules if dataset is empty
        if entropy(data, target) == 0: # dataset is pure, 
            print("otherwise =>",majority_class(data, target))
            data = data.iloc[0:0] # delete all row in data
        else:
            best_entropy = entropy(data,target)
            best_attr = ""
            best_val = ""
            best_data = data
            """
            For every attributes, and every value of that attribute:
                which one yields the least entropy is the best condition to become a rule
            """
            for col in data.columns: 
                vCounts = pd.value_counts(data[col])
                for target_val in vCounts.axes[0]:
                    data2 = data.loc[data[col] == target_val]
                    entropy_data2 = entropy(data2, target)
                    if entropy_data2 < best_entropy:
                        best_entropy = entropy_data2
                        best_attr = col
                        best_val = target_val
                        best_data = data2
            print(best_attr, "=",best_val,"=>",majority_class(best_data, target))
            data = data.loc[data[best_attr] != best_val]

```

```python
simple_rule_learner(sail_df,"Sail")
```

```python
simple_rule_learner(zoo_df,"type")
```
